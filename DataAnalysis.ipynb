Let's inspect the `df` by checking each condition separately to understand why the combined filter resulted in an empty DataFrame.

```python
# Books published in 2019
books_2019 = df[df['Year'] == 2019]
display(books_2019.head())
```

```python
# Books with more than 10,000 reviews
books_high_reviews = df[df['Reviews'] > 10000]
display(books_high_reviews.head())
```

By examining the outputs above, you can see if any specific books appear in both filtered lists. If a book exists in both, then the original `specific_book` filter *should* have returned it. If `specific_book` is still empty, it implies that the intersection of these two conditions is indeed empty for your dataset, or there might be an issue with data types that isn't immediately apparent from `df.info()`.
