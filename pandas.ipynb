```python
data = {

    "First_Name": ["Aman", "Riya", "Kabir", np.nan, "Neel"],
    "Last_Name": ["Sharma", "Gupta", "Khan", np.nan, "Patel"],
    "Age": [21, 19, 22, "NA", 23],
    "Course": ["DS", "AI", "Web", None, "AI"],
    "Marks": [88, 92, 76, None, "Missing"]
}
df = pd.DataFrame(data)

df.replace('NA',np.nan,inplace=True)
df.replace('Missing',np.nan,inplace=True)
df
```
